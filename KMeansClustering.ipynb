```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("iris.csv")
```

```python
data
```

```python
data.columns
```

```python
X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
X
```

```python
y = pd.DataFrame(data["species"].replace({'setosa':0,'versicolor':1,'virginica':2}))
```

```python
y
```

```python
plt.figure(figsize=(14,7))

colormap = np.array(["red","blue","black"])

plt.subplot(1,2,1)
plt.scatter(X.sepal_length,X.sepal_width,c=colormap[y.species])

plt.title("Sepal Data")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")




plt.subplot(1,2,2)
plt.scatter(X.petal_length,X.petal_width,c=colormap[y.species])

plt.title("Petal Data")
plt.xlabel("Petl Length")
plt.ylabel("Petal Width")
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(X)
```

```python
model.labels_
```

```python
plt.figure(figsize=(14,7))

colormap = np.array(["orange","green","crimson"])

plt.subplot(1,2,1)

plt.scatter(X.petal_length,X.petal_width,c=colormap[y.species])
plt.title("Petal Data")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")


plt.subplot(1,2,2)

plt.scatter(X.petal_length,X.petal_width,c=colormap[model.labels_])
plt.title("Petal Model Data")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")

```

```python
plt.figure(figsize=(14,7))

colormap = np.array(["navy","cyan","violet"])

plt.subplot(1,2,1)

plt.scatter(X.sepal_length,X.sepal_width,c=colormap[y.species])
plt.title("Sepal Data")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")


plt.subplot(1,2,2)


plt.scatter(X.sepal_length,X.sepal_width,c=colormap[model.labels_])
plt.title("Sepal Model Data")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```
